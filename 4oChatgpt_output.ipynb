```python
import pandas as pd
import openai

# Set up OpenAI API key
openai.api_key = "sk-proj-"
import pandas as pd
df = pd.read_excel("Pure_600_llama_output.xlsx")
# Define function to generate response using ChatGPT API
def generate_response_model1(user_text):
    try:
        if not user_text:
            return "Error: No input provided."

        print("Generating response for input:", user_text)

        try:
            # Generate response using OpenAI's GPT-3.5 API
            completions = openai.ChatCompletion.create(
                model="gpt-4o",
                messages=[
                    {
                        "role": "user",
                        "content": f'''Please create a bug report that includes the following sections:
1. Steps to Reproduce (S2R): Detailed steps to replicate the issue.
2. Expected Result (ER): What you expected to happen.
3. Actual Result (AR): What actually happened.
4. Additional Information: Include relevant details such as software version, build number, environment, etc.
Highlight the missing information to the reporter if software version, build number, environment, etc not present.
here is raw report {user_text}''',
                    },
                ]
            )

            # Extract the response
            gpt_response = completions['choices'][0]['message']['content']
            print("Generated GPT Response:", gpt_response)

        except Exception as e:
            print("Error generating response:", e)
            gpt_response = "Error generating response."

        return gpt_response

    except Exception as e:
        print("Unexpected error in generate_response_model1:", e)
        return "An unexpected error occurred while processing your request."

# Load the dataset
df = pd.read_excel("CTQRS_200_Score_test_qwen.xlsx")  # Replace with the actual dataset file
df = df[:200]
results = []

for index, row in df.iterrows():
    summary = row['Summary']
    actual_report = row['Actual Report']
    # mistral_repo = row['Mistral Report']
    # pure_llama = row['Output']
    
    # Generate response using ChatGPT API
    generated_response = generate_response_model1(summary)

    # Collect results
    print(f"Processing index {index}")
    results.append({
        'Summary': summary,
        'Actual Report': actual_report,
        # 'Mistral Report': mistral_repo,
        '4o_gpt Output': generated_response,
        # 'Pura llama Output': pure_llama,
    })

# Create DataFrame from results
results_df = pd.DataFrame(results)

# Save results to Excel
results_df.to_excel("Finall_200_4o_GPT_Bug_Reports.xlsx", index=False)

print("Processing complete. Results saved to 'Final_4o_GPT_Bug_Reports.xlsx'.")
```
